## Custom Tools

```python
from langchain.tools import tool
```

```python
# Step 1 - create a function

def multiply(a, b):
    """ Multiply two numbers """
    return a * b
```

```python
# Step 2 - add type hints

def multiply(a : int, b : int ) -> int:
    """ Multiply two numbers """
    return a * b
```

```python
# Step 3 - add tool decorator

@tool
def multiply(a : int, b : int ) -> int:
    """ Multiply two numbers """
    return a * b
```

```python
result = multiply.invoke({"a" : 3, "b" : 5})

print(result)
```

```python
print(multiply.name)
print(multiply.description)
print(multiply.args)
```

```python
print(multiply.args_schema.model_json_schema())
```

## Method 2 - Using Structured Tools

```python
from langchain_community.tools import StructuredTool
from pydantic import BaseModel, Field
```

```python
class MultiplyInput(BaseModel):
    a : int = Field(description="First number to multiply")
    b : int = Field(description="Second number to multiply")
```

```python
def multiply_funct(a : int, b : int ) -> int:
    """ Multiply two numbers """
    return a * b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_funct,
    name= "multiply",
    description="Multiply two numbers",
    args_schema=MultiplyInput

)
```

```python
result = multiply_tool.invoke({'a' : 4, 'b' : 5})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

## Method 3 - Using BaseTool 

```python
from langchain_community.tools import BaseTool
from typing import Type
```

```python
class MultiplyInput(BaseModel):
    a : int = Field(description="First number to multiply")
    b : int = Field(description="Second number to multiply")
```

```python
class MultiplyTool(BaseTool):
    name : str = "multiply"
    description : str = "multiply two numbers"
    args_schema: Type[BaseModel] = MultiplyInput

    def _run(self, a : int, b : int) -> int:
        return a * b
```

```python
multiply_tool = MultiplyTool()
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)

print(multiply_tool.args)
```

## Toolkits

```python
from langchain_community.tools import tool
```

```python
@tool
def add(a :int, b : int) -> int:
    """ Add two numbers """
    return a + b

@tool
def multiply(a :int, b : int) -> int:
    """ Multiply two numbers """
    return a * b
```

```python
class MathToolkit:
    def get_tools(self):
        return [add, multiply]
    
```

```python
toolkit = MathToolkit()

tools = toolkit.get_tools()

for tool in tools:
    print(tool.name, " => ", tool.description)
```
